```python
movies.dropna(inplace=True)
```

```python
movies.iloc[0].genres
```

```python
#Converting string to list

ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
#Function to convert list of dictionaries to list format

def convert(text):
  L = []
  for i in ast.literal_eval(text):
    L.append(i['name']) 
  return L 
```

```python
movies['genres'] = movies['genres'].apply(convert)
movies['keywords'] = movies['keywords'].apply(convert)
movies.head()
```

```python
movies['cast'][0]
```

```python
# Function that will return 3 actor names from a list of dictionaries

def convert3(text):
  L = []
  counter = 0
  for i in ast.literal_eval(text):
    if counter != 3:
      L.append(i['name']) 
    counter += 1
  return L 
```

```python
movies['cast'] = movies['cast'].apply(convert3)
movies.head()
```

```python
# Function that will return the name of the Director from a list of dictionaries

def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L 
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
# Changing type of Overview from list to string

movies['overview'] = movies['overview'].apply(lambda x:x.split())
movies.head()
```

```python
# Using lambda function to remove all blank spaces. This is done so that there is no confusion in names when tags are made.

movies['genres'] = movies['genres'].apply(lambda x: [i.replace(" ","") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x: [i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply(lambda x: [i.replace(" ","") for i in x])
movies['crew'] = movies['crew'].apply(lambda x: [i.replace(" ","") for i in x])

movies.head()
```

```python
# Creating a new column- TAGS, that contains words from all other columns

movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies.head()
```

```python
# Creating a new dataframe with the following 3 columns

new_df = movies[['movie_id', 'title', 'tags']]
```

```python
new_df
```

```python
# Concatinating the list of strings in Tags to a single string

new_df['tags'] = new_df['tags'].apply(lambda x: " ".join(x))
new_df.head()
```

```python
new_df['tags'][0]
```

```python
# Converting to lowercase

new_df['tags'] = new_df['tags'].apply(lambda x: x.lower())
new_df.head()
```

## Stemming

```python
# Applying Stemming to the corpus

# E.g: ['loved', 'loving', 'love'] = ['love', 'love', 'love']
```

```python
import nltk
from nltk.stem.porter import PorterStemmer

ps = PorterStemmer() # Creating a PorterStemmer object
```

```python
def stem(text):
  y = []

  for i in text.split():
    y.append(ps.stem(i))
  
  return " ".join(y)
```

```python
# Applying stemming to all tags in the corpus

new_df['tags'] = new_df['tags'].apply(stem)
```

```python
new_df['tags'][0]
```

# Vectorization

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5000, stop_words = 'english')
```

# Recommender System

```python
# 1. Given a movie, first find index i of that movie (0 to 4805) and take similarity[i] array
# 2. Sort the array so that most similiar movies appear first and least similar appear last
# 3. Return top 5/10/... movies as per choice
```

```python
new_df[new_df['title'] == 'Batman Begins'].index[0]
```

```python
# To get a list of tuples that contains (index of movie, similarity of index movie with movie 0)

list(enumerate(similarity[0]))
```

```python
# Sorting based on 2nd element in tuple- (similarity of index movie with movie 0)
# Then selecting the first 5 movies (Excluding movie 0- the movie itself)

sorted(list(enumerate(similarity[0])), reverse = True, key = lambda x: x[1])[1:6]
```

```python
new_df.iloc[1216].title
```

```python
# Using the above in the function below

def recommend(movie):
  movie_index = new_df[new_df['title'] == movie].index[0]
  distances = similarity[movie_index]
  movies_list = sorted(list(enumerate(distances)), reverse = True, key = lambda x: x[1])[1:6]

  for i in movies_list:
    print(new_df.iloc[i[0]].title)
```

```python
recommend('Avatar')
```
